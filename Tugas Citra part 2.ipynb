```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Downloads/thewalloflove.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Copy the image
img_copy = img.copy()

# Draw a rectangle 
cv2.rectangle(img_copy, pt1 = (800, 470), pt2 = (980, 530), color = (255, 0, 0), thickness = 5)
plt.imshow(img_copy)
```

```python
# Draw a circle 
cv2.circle(img_copy, center = (950, 50), radius = 50, color = (0, 0, 255), thickness = 5)
plt.imgshow(img_copy)
```

```python
# Draw a circle 
cv2.circle(img_copy, center = (950, 50), radius = 50, color = (0, 0, 255), thickness = 5)
plt.imshow(img_copy)
```

```python
cv2.putText(img_copy, text = "the Wall of Love", org = (250, 250), fontFace = cv2.FONT_HERSHEY_DUPLEX,
            fontScale = 2, color = (0, 255, 0), thickness = 2, lineType = cv2.LINE_AA)
plt.imshow(img_copy)
```

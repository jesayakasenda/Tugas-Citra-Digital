```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Downloads/images.jpg')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvrColor(img, cv2.COLOR_BGR2GRAY)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Plot the three channels of the image 
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
```

```python
# Plot the three channels of the image 
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray') 
plt.show()
```

```python
# Transform the image into HSV and HLS models 
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) 
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
```

```python
# Plot the converted images 
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20)) 
ax1.imshow(img_hsv) 
ax2.imshow(img_hls) 
plt.show()
```

```python
# Draw a rectangle 
cv2.rectangle(img_copy, pt1 = (800, 470), pt2 = (980, 530), color = (255, 0, 0), thickness = 5) 
plt.imshow(img_copy)
```

```python
# Draw a circle 
cv2.circle(img_copy, center = (950, 50), radius = 50, color = (0, 0, 255), thickness = 5) 
plt.imshow(img_copy)
```

```python
# Add text 
cv2.putText(img_copy, text = "the Wall of Love",             
            org = (250, 250),            
            fontFace = cv2.FONT_HERSHEY_DUPLEX,             
            fontScale = 2,             
            color = (0, 255, 0),             
            thickness = 2,             
            lineType = cv2.LINE_AA) 
plt.imshow(img_copy)
```
